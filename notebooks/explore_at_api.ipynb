```python
import os
import requests

API_KEY = os.getenv("AT_API_KEY")
url = "https://api.at.govt.nz/realtime/legacy/vehiclelocations"
headers = {"Ocp-Apim-Subscription-Key": API_KEY}

response = requests.get(url, headers=headers)
print(response.status_code)
print(response.content[:200]) # 先看前200字节
```

```python
data = response.json()

# 看顶层结构
print(data.keys())
print(data["response"]["header"])

# 看entity数量和第一条记录
entities = data["response"]["entity"]
print(f"共 {len(entities)} 辆车")
print(entities[0])
```
